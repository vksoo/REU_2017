```python
%matplotlib inline
from __future__ import print_function
import matplotlib.pyplot as plt
import numpy as np
```

```python
freq_lw, s31_wmag = np.genfromtxt('7s31_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -2))
freq_lwo, s31_womag = np.genfromtxt('8s31_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -2))
freq_pw, s31_wphase = np.genfromtxt('7s31_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -1))
freq_pwo, s31_wophase = np.genfromtxt('8s31_phase.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -1))
```

```python
s31_wmag_db = 20. * np.log10(s31_wmag)
s31_womag_db = 20. * np.log10(s31_womag)
```

```python
s31_wphase_r = np.radians(s31_wphase)
s31_wophase_r = np.radians(s31_wophase)
```

```python
fig = plt.figure(figsize = (14., 10.))

axes1 = fig.add_subplot(2,2, 1)
axes2 = fig.add_subplot(2, 2 ,2)
axes3 = fig.add_subplot(2, 2 ,3)
axes4 = fig.add_subplot(2, 2 ,4)

axes1.set_title('$S31$ $Amplitude$ With Washers')
axes1.set_ylabel('$S31$ $Log$ $Magnitude$ $(dB)$')
axes1.set_xlabel('$Frequency$ $(MHz)$')
axes1.plot(freq_lw, s31_wmag_db)

axes2.set_title('$S31$ $Phase$ With Washers')
axes2.set_ylabel('$Phase$ $(Radians)$')
axes2.set_xlabel('$Frequency$ $(MHz)$')
axes2.plot(freq_pw, s31_wphase_r)

axes3.set_title('$S31$ $Amplitude$ Without Washers')
axes3.set_ylabel('$S31$ $Log$ $Magnitude$ $(dB)$')
axes3.set_xlabel('$Frequency$ $(MHz)$')
axes3.plot(freq_lwo, s31_womag_db)

axes4.set_title('$S31$ $Phase$ Without Washers')
axes4.set_ylabel('$Phase$ $(Radians)$')
axes4.set_xlabel('$Frequency$ $(MHz)$')
axes4.plot(freq_pwo, s31_wophase_r)

axes1.set_ylim([-4.2, -3.60])
axes3.set_ylim([-4.2, -3.60])
axes2.set_ylim([-1.8, -0.2])
axes4.set_ylim([-1.8, -0.2])

fig.tight_layout()
```
