```python
%matplotlib inline
from __future__ import print_function
import matplotlib.pyplot as plt
import numpy as np
```

```python
freq_lw, s11_wmag = np.genfromtxt('7s11_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -4))
freq_lwo, s11_womag = np.genfromtxt('8s11_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -4))
freq_pw, s11_wphase = np.genfromtxt('7s11_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -3))
freq_pwo, s11_wophase = np.genfromtxt('8s11_phase.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -3))
```

```python
s11_wmag_db = 20. * np.log10(s11_wmag)
s11_womag_db = 20. * np.log10(s11_womag)
```

```python
s11_wphase_r = np.radians(s11_wphase)
s11_wophase_r = np.radians(s11_wophase)
```

```python
fig = plt.figure(figsize = (14., 10.))

axes1 = fig.add_subplot(2,2, 1)
axes2 = fig.add_subplot(2, 2 ,2)
axes3 = fig.add_subplot(2, 2 ,3)
axes4 = fig.add_subplot(2, 2 ,4)

axes1.set_title('$S11$ $Amplitude$ With Washers')
axes1.set_ylabel('$S11$ $Log$ $Magnitude$ $(dB)$')
axes1.set_xlabel('$Frequency$ $(MHz)$')
axes1.plot(freq_lw, s11_wmag_db)

axes2.set_title('$S11$ $Phase$ With Washers')
axes2.set_ylabel('$Phase$ $(Radians)$')
axes2.set_xlabel('$Frequency$ $(MHz)$')
axes2.plot(freq_pw, s11_wphase_r)

axes3.set_title('$S11$ $Amplitude$ Without Washers')
axes3.set_ylabel('$S11$ $Log$ $Magnitude$ $(dB)$')
axes3.set_xlabel('$Frequency$ $(MHz)$')
axes3.plot(freq_lwo, s11_womag_db)

axes4.set_title('$S11$ $Phase$ Without Washers')
axes4.set_ylabel('$Phase$ $(Radians)$')
axes4.set_xlabel('$Frequency$ $(MHz)$')
axes4.plot(freq_pwo, s11_wophase_r)

axes1.set_ylim([-14.0, -5.0])
axes3.set_ylim([-14.0, -5.0])
axes2.set_ylim([-3.50, 3.50])
axes4.set_ylim([-3.50, 3.50])


fig.tight_layout()
```
