```python
%matplotlib inline
from __future__ import print_function
import matplotlib.pyplot as plt
import numpy as np
```

```python
freq_lw, s21_wmag = np.genfromtxt('7s21_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -2))
freq_lwo, s21_womag = np.genfromtxt('8s21_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -2))
freq_pw, s21_wphase = np.genfromtxt('7s21_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -1))
freq_pwo, s21_wophase = np.genfromtxt('8s21_phase.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -1))
```

```python
s21_wmag_db = 20. * np.log10(s21_wmag)
s21_womag_db = 20. * np.log10(s21_womag)
```

```python
s21_wphase_r = np.radians(s21_wphase)
s21_wophase_r = np.radians(s21_wophase)
```

```python
fig = plt.figure(figsize = (14., 10.))

axes1 = fig.add_subplot(2,2, 1)
axes2 = fig.add_subplot(2, 2 ,2)
axes3 = fig.add_subplot(2, 2 ,3)
axes4 = fig.add_subplot(2, 2 ,4)

axes1.set_title('$S21$ $Amplitude$ With Washers')
axes1.set_ylabel('$S21$ $Log$ $Magnitude$ $(dB)$')
axes1.set_xlabel('$Frequency$ $(MHz)$')
axes1.plot(freq_lw, s21_wmag_db)

axes2.set_title('$S21$ $Phase$ With Washers')
axes2.set_ylabel('$Phase$ $(Radians)$')
axes2.set_xlabel('$Frequency$ $(MHz)$')
axes2.plot(freq_pw, s21_wphase_r)

axes3.set_title('$S21$ $Amplitude$ Without Washers')
axes3.set_ylabel('$S21$ $Log$ $Magnitude$ $(dB)$')
axes3.set_xlabel('$Frequency$ $(MHz)$')
axes3.plot(freq_lwo, s21_womag_db)

axes4.set_title('$S21$ $Phase$ Without Washers')
axes4.set_ylabel('$Phase$ $(Radians)$')
axes4.set_xlabel('$Frequency$ $(MHz)$')
axes4.plot(freq_pwo, s21_wophase_r)

axes1.set_ylim([-14., -3.])
axes3.set_ylim([-14., -3.])
axes2.set_ylim([-3.5, 3.5])
axes4.set_ylim([-3.5, 3.5])

fig.tight_layout()
```
