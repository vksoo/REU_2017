```python
%matplotlib inline
from __future__ import print_function
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
from IPython.display import Image
```

```python
s12_freq_p_0, s12_phase_0 = np.genfromtxt('0s12_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -1))
s12_freq_l_0, s12_mag_0 = np.genfromtxt('0s12_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -2))
s22_freq_p_0, s22_phase_0 = np.genfromtxt('0s22_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -3))
s22_freq_l_0, s22_mag_0 = np.genfromtxt('0s22_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -4))

s12_freq_p_1, s12_phase_1 = np.genfromtxt('1s12_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -1))
s12_freq_l_1, s12_mag_1 = np.genfromtxt('1s12_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -2))
s22_freq_p_1, s22_phase_1 = np.genfromtxt('1s22_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -3))
s22_freq_l_1, s22_mag_1 = np.genfromtxt('1s22_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -4))

s12_freq_p_2, s12_phase_2 = np.genfromtxt('2s12_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -1))
s12_freq_l_2, s12_mag_2 = np.genfromtxt('2s12_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -2))
s22_freq_p_2, s22_phase_2 = np.genfromtxt('2s22_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -3))
s22_freq_l_2, s22_mag_2 = np.genfromtxt('2s22_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -4))

s12_freq_p_3, s12_phase_3 = np.genfromtxt('3s12_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -1))
s12_freq_l_3, s12_mag_3 = np.genfromtxt('3s12_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -2))
s22_freq_p_3, s22_phase_3 = np.genfromtxt('3s22_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -3))
s22_freq_l_3, s22_mag_3 = np.genfromtxt('3s22_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -4))

s12_freq_p_4, s12_phase_4 = np.genfromtxt('4s12_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -1))
s12_freq_l_4, s12_mag_4 = np.genfromtxt('4s12_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -2))
s22_freq_p_4, s22_phase_4 = np.genfromtxt('4s22_phase.csv', delimiter = ',', unpack = True, skip_header= 305, skip_footer = 0, usecols = (5, -3))
s22_freq_l_4, s22_mag_4 = np.genfromtxt('4s22_log.csv', delimiter = ',', unpack = True, skip_header = 305, skip_footer = 0, usecols = (5, -4))
```

```python
#Convert to decibels
s12_mag_db_0 = 20. * np.log10(s12_mag_0)
s22_mag_db_0 = 20. * np.log10(s22_mag_0)

s12_mag_db_1 = 20. * np.log10(s12_mag_1)
s22_mag_db_1 = 20. * np.log10(s22_mag_1)

s12_mag_db_2 = 20. * np.log10(s12_mag_2)
s22_mag_db_2 = 20. * np.log10(s22_mag_2)

s12_mag_db_3 = 20. * np.log10(s12_mag_3)
s22_mag_db_3 = 20. * np.log10(s22_mag_3)

s12_mag_db_4 = 20. * np.log10(s12_mag_4)
s22_mag_db_4 = 20. * np.log10(s22_mag_4)
```

```python
#Convert to radians
s12_phase_r_0 = np.radians(s12_phase_0)
s22_phase_r_0 = np.radians(s22_phase_0)

s12_phase_r_1 = np.radians(s12_phase_1)
s22_phase_r_1 = np.radians(s22_phase_1)

s12_phase_r_2 = np.radians(s12_phase_2)
s22_phase_r_2 = np.radians(s22_phase_2)

s12_phase_r_3 = np.radians(s12_phase_3)
s22_phase_r_3 = np.radians(s22_phase_3)

s12_phase_r_4 = np.radians(s12_phase_4)
s22_phase_r_4 = np.radians(s22_phase_4)
```

```python
#Plot of Amplitude for 0 to 4 screws data on this single plot.
fig_1 = figure(figsize = (16,8))

plot(s12_freq_l_0, s12_mag_db_0, 'y-', label = 'Zero screws')
plot(s12_freq_l_1, s12_mag_db_1, 'b-', label = 'One screw')
plot(s12_freq_l_2, s12_mag_db_2, 'r-', label = 'Two screws')
plot(s12_freq_l_3, s12_mag_db_3, 'g-', label = 'Three screws')
plot(s12_freq_l_4, s12_mag_db_4, 'm-', label = 'Four screws')

legend(loc = 'best')
ylabel('$S12 \, Log \, Magnitude \, \mathrm{(dB)}$', fontsize = 12) # {space}\, is for spacing
xlabel('$Frequency \, \mathrm{(MHz)}$', fontsize = 12)

ylim(-27.5, -5) #first value to -12.5 for close up
xlim(50, 250)
grid('on')
title('S12 Amplitude', fontsize = 16).set_position([0.5,1.02])
#savefig('S12_amplitude.pdf')
```

```python
fig_2 = figure(figsize = (16,8))

plot(s12_freq_p_0, s12_phase_r_0, 'y-', label = 'Zero screws')
plot(s12_freq_p_1, s12_phase_r_1, 'b-', label = 'One screw')
plot(s12_freq_p_2, s12_phase_r_2, 'r-', label = 'Two screws')
plot(s12_freq_p_3, s12_phase_r_3, 'g-', label = 'Three screws')
plot(s12_freq_p_4, s12_phase_r_4, 'm-', label = 'Four screws')

#This is to compare the S12 and S22
'''
plot(s22_freq_p_0, s22_phase_r_0)
plot(s22_freq_p_1, s22_phase_r_1)
plot(s22_freq_p_2, s22_phase_r_2)
plot(s22_freq_p_3, s22_phase_r_3)
plot(s22_freq_p_4, s22_phase_r_4)
'''

legend(loc = 'best')
xlabel('$Frequency \, \mathrm{(MHz)}$', fontsize = 12)
ylabel(r'S12 Phase, $\varphi$,  (radians)', fontsize = 12)

ylim(-4, 4)
xlim(50, 250)
grid('on')
title('S12 Phase', fontsize = 16).set_position([0.5,1.02])
#savefig('S12_phase.pdf')
```

```python
fig_3 = figure(figsize = (16,8))

plot(s22_freq_l_0, s22_mag_db_0, 'y-', label = 'Zero screws')
plot(s22_freq_l_1, s22_mag_db_1, 'b-', label = 'One screw')
plot(s22_freq_l_2, s22_mag_db_2, 'r-', label = 'Two screws')
plot(s22_freq_l_3, s22_mag_db_3, 'g-', label = 'Three screws')
plot(s22_freq_l_4, s22_mag_db_4, 'm-', label = 'Four screws')

legend(loc = 'best')
ylabel('$S22 \, Log \, Magnitude \, \mathrm{(dB)}$', fontsize = 12)
xlabel('$Frequency \, \mathrm{(MHz)}$', fontsize = 12)

#ylim(-27.5, -5)
xlim(50, 250)
grid('on')
title('S22 Amplitude', fontsize = 16).set_position([0.5,1.02])
#savefig('S22_amplitude.pdf')
```

```python
fig_4 = figure(figsize = (16,8))

plot(s22_freq_p_0, s22_phase_r_0, 'y-', label = 'Zero screws')
plot(s22_freq_p_1, s22_phase_r_1, 'b-', label = 'One screw')
plot(s22_freq_p_2, s22_phase_r_2, 'r-', label = 'Two screws')
plot(s22_freq_p_3, s22_phase_r_3, 'g-', label = 'Three screws')
plot(s22_freq_p_4, s22_phase_r_4, 'm-', label = 'Four screws')

legend(loc = 'best')
xlabel('$Frequency \, \mathrm{(MHz)}$', fontsize = 12)
ylabel(r'S22 Phase, $\varphi$, (radians)', fontsize = 12) #rawstring with r'string' for \varphi;not needed for \phi

ylim(-4, 4)
xlim(50, 250)
grid('on')
title('S22 Phase', fontsize = 16).set_position([0.5,1.02])
#savefig('S22_phase.pdf')
```
