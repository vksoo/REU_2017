```python
#S-parameters as s_**
s_11 = s11
s_1c = (s12 + s13) / np.sqrt(2.)
s_1d = (s12 - s13) / np.sqrt(2.)

s_c1 = (s21 + s31) / np.sqrt(2.)
s_cc = (1. / 2.) * (s22 + s23 + s32 + s33)
s_cd = (1. / 2.) * (s22 - s23 + s32 - s33)

s_d1 = (s21 - s31) / np.sqrt(2.)
s_dc = (1. / 2.) * (s22 + s23 - s32 - s33)
s_dd = (1. / 2.) * (s22 - s23 - s32 + s33)
```
